```python
import pandas as pd
```

```python
data = pd.read_csv('pre_analysis_last.csv')
data = data.rename(columns={"AC": "IC"})
```

```python
turkish_data = data[data['is_turkish']== 1] 
international_data = data[data['is_turkish']== 0] 
donor = pd.read_csv('donor.csv')
donor = donor.rename(columns={"AC": "IC"})
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Veriyi hazırla
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

# Her bir metrik için (CH, AF, IC)
metrics = ['CH', 'AF', 'IC']
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']

# Hastane isimlerini al
hospitals = donor['Hospital'].unique()

for idx, (ax, metric, color) in enumerate(zip(axes, metrics, colors)):
    # Her hastane için scatter plot
    for hospital in hospitals:
        hospital_data = donor[donor['Hospital'] == hospital]
        
        for _, row in hospital_data.iterrows():
            # Y ekseninden noktaya çizgi çek
            ax.plot([0, row[metric]], [hospital, hospital], 
                   color='gray', linewidth=2, alpha=0.5, zorder=1)
            
            # Noktayı çiz
            ax.scatter(row[metric], hospital,
                      s=row['log_embryo_count'] * 100,
                      alpha=0.7,
                      color=color,
                      edgecolors='black',
                      linewidth=1.5,
                      zorder=2)
    
    ax.set_xlabel(metric, fontsize=12, fontweight='bold')
    ax.set_title(f'{metric} Değerleri', fontsize=11)
    ax.grid(True, alpha=0.3, axis='x', linestyle='--')
    
    if idx == 0:
        ax.set_ylabel('Hastane', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.suptitle('Donor Verileri - Hastane Bazlı Dotplot (Nokta Boyutu: Embryo Count)', 
             fontsize=14, fontweight='bold', y=1.02)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Ensure we don't modify original unexpectedly
international = international_data.copy()

# (optionally) drop the specific rows you mentioned before
international = international.drop([0,1,2,3,21], errors='ignore')

# create log and groups (4 quantile-based groups: Very Low, Low, Medium, High)
international['log_embryo_count'] = np.log(international['embryo_count'] + 1)
international['embryo_group'] = pd.qcut(international['embryo_count'].fillna(0), 4, labels=['Very Low','Low','Medium','High'])

metrics = ['CH', 'AF', 'IC']
groups = ['Very Low', 'Low', 'Medium', 'High']

# Figure with rows = groups, cols = metrics (increased height)
fig, axes = plt.subplots(len(groups), len(metrics), figsize=(22, 6 * len(groups)), sharex='col')

for i, group in enumerate(groups):
    dfg = international[international['embryo_group'] == group]
    if dfg.empty:
        # empty row: turn off axes
        for j in range(len(metrics)):
            axes[i, j].axis('off')
        continue

    # hospitals in this group and numeric y positions
    hospitals = dfg['Hospital'].unique()
    y_positions = np.arange(len(hospitals))
    mapping = {h: y for h, y in zip(hospitals, y_positions)}
    dfg = dfg.assign(_y = dfg['Hospital'].map(mapping))

    # normalize sizes inside group for visibility
    max_emb = dfg['embryo_count'].replace(0, np.nan).max()
    if np.isnan(max_emb) or max_emb == 0:
        sizes = np.full(len(dfg), 30)
    else:
        sizes = (dfg['embryo_count'] / max_emb) * 300 + 20

    for j, metric in enumerate(metrics):
        ax = axes[i, j]
        # draw horizontal lines from x=0 to metric value
        ax.hlines(dfg['_y'], xmin=0, xmax=dfg[metric], color='gray', linewidth=1.6, alpha=0.6, zorder=1)

        # scatter points
        ax.scatter(dfg[metric], dfg['_y'], s=sizes, alpha=0.8,
                   color=['#FF6B6B', '#4ECDC4', '#45B7D1'][j],
                   edgecolors='black', linewidth=0.8, zorder=2)

        ax.set_xlabel(metric, fontsize=11, fontweight='bold')
        ax.grid(True, axis='x', linestyle='--', alpha=0.3)

        # show hospital labels only in first column to reduce clutter
        if j == 0:
            ax.set_yticks(y_positions)
            ax.set_yticklabels(hospitals, fontsize=8)
            ax.set_ylabel(f'{group} group\nHospital', fontsize=10, fontweight='bold')
        else:
            ax.set_yticks([])

        ax.set_title(f'{group} — {metric}', fontsize=11)

plt.suptitle('International Data — Grouped by Embryo Count (Very Low / Low / Medium / High)', fontsize=14, fontweight='bold', y=1.0)
plt.tight_layout()
plt.show()
```
